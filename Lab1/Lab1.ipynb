# Pengenalan Pandas

[__Pandas__](https://pandas.pydata.org/docs/) adalah _library_ untuk bahasa Python yang menyediakan struktur data dan fungsi-fungsi yang mempermudah pemrosesan dan analisis data.

```python
# import library
import pandas as pd
```

## Objek _DataFrame_

_DataFrame_ adalah struktur data utama pada Pandas. Struktur data ini dapat digunakan untuk menyimpan data tabel 2 dimensi. Baris suatu _DataFrame_ terurut mulai dari 0. Jika nama-nama kolom tidak diberikan, maka kolom juga akan terurut dari 0.

```python
# Membuat DataFrame dari collection generik

lst = [
    [1,2,3],
    ['a','b','c'],
    [None, 1.0, 2.0]
]
df1 = pd.DataFrame(lst)
display(df1) # Menampilkan DataFrame
```

```python
# Membuat DataFrame dengan membaca data dari suatu berkas

df2 = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', # Path menuju berkas dataset, dapat berupa path lokal maupun remote
    header=None # Baris pertama dataset bukanlah header
)
display(df2) # Menampilkan DataFrame
```

Memberikan nama kolom pada _DataFrame_ agar lebih bermakna

```python
df2.columns = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'label']
df2
```

## Mengakses baris dan kolom _DataFrame_

Kita dapat mengakses kolom _DataFrame_ seperti mengakses indeks pada array atau list, baik sebagai Series maupun DataFrame baru.

```python
# Mengakses kolom sebagai Series
df2_sepal_len = df2['sepal-length']
print(type(df2_sepal_len))
df2_sepal_len
```

```python
# Mengakses kolom sebagai DataFrame
df2_sepal_len_2 = df2[['sepal-length']]
print(type(df2_sepal_len_2))
df2_sepal_len_2
```

```python
# Mengakses lebih dari 1 kolom sebagai DataFrame
df2_sepal = df2[['sepal-length', 'sepal-width']]
df2_sepal
```

Akses terhadap __tepat satu__ kolom juga dapat dilakukan dengan operator __dot__ (.) dengan syarat kolom memiliki nama (tidak menggunakan indeks bilangan _default_). Nama kolom tidak boleh menggunakan nama atribut dari objek _DataFrame_. Nama kolom yang dapat diakses dengan operator ini juga tidak boleh mengandung karakter-karakter yang dilarang dalam nama variabel Python. Perhatikan contoh berikut.

```python
# Mengakses kolom dengan nama yang baik
display(df2.label)

print('\n')

# Mengakses kolom dengan nama yang kurang baik
try:
    display(df2.sepal-length)
except Exception as e:
    print('Terjadi exception:', e)
    print('Error terjadi karena "df2.sepal-length" dianggap tiga token terpisah, yaitu "df2.sepal", "-", dan "length".')
```

Kita dapat mengakses baris (dan kolom) _DataFrame_ menggunakan <code>__loc__</code> atau <code>__iloc__</code>. <code>loc</code> mengakses indeks berdasarkan label dari baris atau kolom, sedangkan <code>iloc</code> mengakses indeks berdasarkan posisi dalam bilangan bulat. <code>loc</code> juga menerima _range_ secara inklusif.

```python
# Mengakses baris 0 sampai dengan 3 menggunakan loc
df2.loc[0:3]
```

```python
# Mengakses baris 0 sampai dengan 3 menggunakan iloc
df2.iloc[0:4]
```

```python
# Mengakses baris 0 - 3, namun hanya kolom petal saja, menggunakan loc
df2.loc[0:3, ["petal-length", "petal-width"]]
```

```python
# Mengakses baris 0 - 3, namun hanya kolom petal saja, menggunakan iloc
df2.iloc[0:4, 2:4]
```

```python
# Memasukkan indeks spesifik menggunakan loc dan iloc
print(df2.loc[0,"class"])
print(df2.iloc[0,4])
```

_Indexing_ juga dapat dilakukan tanpa menggunakan <code>loc</code> maupun <code>iloc</code>, dengan urutan indeks dimulai dari kolom, kemudian indeks dari baris.

```python
# Indexing tanpa menggunakan loc maupun iloc
print(df2["class"][0])
```

## Menangani _Missing Values_

Nilai _null_ atau "_missing values_" adalah informasi yang diisi dengan _null_ atau tidak diisi. Hal ini dapat terjadi karena informasi tersebut memang tidak diberikan (namun ada), sulit dicari, atau memang tidak ada. _Missing values_ juga dapat tersimpan sebagai "NA" atau "NaN".

```python
# Menampilkan jumlah missing values tiap kolom
print("Count missing values untuk df1\n", df1.isnull().sum())
print("\n")
print("Count missing values untuk df2\n", df2.isnull().sum())
```

Dapat dilihat di atas bahwa pada df1 ada 1 nilai _null_ pada kolom 0, sedangkan pada df2 (dataset iris) tidak ditemukan nilai _null_. Selanjutnya kita coba membaca dataset baru pada berkas '<code>MELBOURNE_HOUSE_PRICES_LITE.csv</code>' yang dilampirkan di SCELE.

```python
df3 = pd.read_csv('MELBOURNE_HOUSE_PRICES_LITE.csv')

# Menampilkan jumlah missing values tiap kolom
print("Count missing values untuk df3\n", df3.isnull().sum())
```

Dapat dilihat di atas bahwa ada kolom "Price" yang memiliki nilai _null_. Hal tersebut ditandai sebagai "NaN" pada _DataFrame_ jika ditampilkan seperti di bawah.

```python
# Menampilkan DataFrame df3
df3.head(10)
```

Ada beberapa pendekatan untuk menangani _missing values_. Pendekatan yang umum digunakan adalah mengisi _missing values_ dengan nilai tertentu menggunakan _method_ <code>fillna()</code> atau menghapus (_drop_) baris atau kolom yang mengandung _missing values_ menggunakan _method_ <code>dropna()</code>. Contoh-contoh di bawah akan berfokus pada kolom "Price".

```python
# fillna dengan nilai rata-rata
mean = df3.Price.mean()
df3_fill_mean = df3.copy()
df3_fill_mean.Price.fillna(mean,inplace=True)
df3_fill_mean.head(10)
```

```python
# fillna dengan nilai max
maxx = df3.Price.max()
df3_fill_max = df3.copy()
df3_fill_max.Price.fillna(maxx,inplace=True)
df3_fill_max.head(10)
```

```python
# dropna untuk menghapus baris yang mengandung missing values
df3_drop_row = df3.dropna()
df3_drop_row.head(10)
```

```python
# dropna untuk menghapus kolom yang mengandung missing values
df3_drop_col = df3.dropna(axis='columns')
df3_drop_col.head(10)
```
